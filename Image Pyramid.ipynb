## Image Pyramid:
An image pyramid is a collection of images - all arising from a single original image - that are successively downsampled until some desired stopping point is reached. Image Pyramid is used to downsample or upsample the images.

There are two common kinds of image pyramids:
1. **Gaussian pyramid**: Used to downsample images
2. **Laplacian pyramid**: Used to reconstruct an upsampled image from an image lower in the pyramid (with less resolution)

### pyrDown:
Blurs an image and downsamples it.Blurring is done using Gaussian Filter.

* First, it convolves the source image with the Gaussian kernel.
* Then, it downsamples the image by rejecting even rows and columns.

```python
# importing library
import cv2
import matplotlib.pyplot as plt
#plt.rcParams['figure.facecolor'] = "#ffffff"
# reading an image
image= cv2.imread("./data/dog.png")

# convertin from BGR to RGB
image_rgb= cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# downscaling the image using cv2.pyrDown 
downsample_img= cv2.pyrDown(src= image_rgb)

#plotting
cv2.imshow("Original Image,shape: {}".format(image_rgb.shape),image_rgb)
cv2.imshow("PyrDown Image, shape: {}".format(downsample_img.shape), downsample_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# we can still downsample the image by providing the downsample_img further into cv2.pyrDown function
downsample_final= cv2.pyrDown(src= downsample_img)

cv2.imshow("Original Image,shape: {}".format(image_rgb.shape),image_rgb)
cv2.imshow("PyrDown Image, shape: {}".format(downsample_img.shape), downsample_img)
cv2.imshow("Further downsampled Image, shape: {}".format(downsample_final.shape), downsample_final)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### pyrUp:
Upsamples an image and then blurs it.

The function performs the upsampling step of the Gaussian pyramid construction, though it can actually be used to construct the Laplacian pyramid. First, it upsamples the source image by injecting even zero rows and columns and then convolves the result with the same kernel as in pyrDown multiplied by 4.

```python
# here lets take the finally downsampled image form above as the input to pyrUp fucntion and see

# importing library
import cv2
import matplotlib.pyplot as plt


# upsampling the image using cv2.pyrDown 
upsample_img= cv2.pyrUp(src= downsample_final)

#plotting
cv2.imshow("Original Image,shape: {}".format(downsample_final.shape),downsample_final)
cv2.imshow("pyrUpward Image, shape: {}".format(upsample_img.shape), upsample_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### pyrMeanShiftFiltering:
Performs initial step of meanshift segmentation of an image.

sp-	The spatial window radius.             
sr-	The color window radius.                    
maxLevel-	Maximum level of the pyramid for the segmentation.                    
termcrit-	Termination criteria: when to stop meanshift iterations.           


Link: https://stackoverflow.com/questions/41558341/what-does-the-color-window-radius-in-pyrmeanshiftfiltering-mean

```python
# importing library
import cv2
import matplotlib.pyplot as plt

# reading an image
image= cv2.imread("./data/dog.png")

# convertin from BGR to RGB
image_rgb= cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# mean shift pyramid algorithm
pyrMeanshift_img= cv2.pyrMeanShiftFiltering(src= image_rgb, sp= 3,sr= 3, maxLevel= 5)


# plotting the imag

cv2.imshow("Original Image,shape: {}".format(image_rgb.shape),image_rgb)
cv2.imshow("Mean Shift Filtering Image, shape: {}".format(pyrMeanshift_img.shape), pyrMeanshift_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
