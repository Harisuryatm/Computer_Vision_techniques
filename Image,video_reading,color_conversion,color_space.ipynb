## Image Processing using Opencv

### Image Reading:

```python
# Reading an image from the folder
#importing libraries
import cv2

# In the case of color images, the decoded images will have the channels stored in B G R order.

dog_im_bgr = cv2.imread(filename= "../computer_vision/OpenCV/images/dog.jpg")

#converting BGR to RGB format
dog_im_rgb = cv2.cvtColor(dog_im_bgr, cv2.COLOR_BGR2RGB)

print("Image size: ,",dog_im_rgb.shape)
print(dog_im_rgb)
```

### Image writing

```python
import cv2

# writing image to the specific folder
cv2.imwrite(filename= './data/dog.png', img= dog_im_bgr)
```

### Image visualization

```python
import cv2
#Using opencv in a new window : automatically converts to RGB format and visulaize
cv2.imshow("Dog",dog_im_bgr)
cv2.waitKey(0) 
```

```python
#Using matplotlib
import matplotlib.pyplot as plt
plt.imshow(dog_im_rgb)
plt.show()
```

### Reading Videos from the camera                   

Link : https://learnopencv.com/read-write-and-display-a-video-using-opencv-cpp-python/

```python
# Reading video from the camera
import cv2
```

```python
#A frame of a video is simply an image and we display each frame the same way we display images, 
#i.e., we use the function imshow().
cap = cv2.VideoCapture(0)  # 0 represent first camera . as we have only one camera

if cap.isOpened() == False:
    print("Error opening the camera/ video stream")

while cap.isOpened():
    # capture frame by frame
    ret, frame= cap.read()
    
    if ret == True:
        # display the frames
        cv2.imshow("Frame", frame)
        
        # if 'q' keypad is touched, it stoped
        if cv2.waitKey(25) & 0XFF == ord('q'):
            break
    
    else:
        break

# when everything done, release the video capturing object
cap.release()

# destroy all the windows
cv2.destroyAllWindows()
```

### Color Space

#### 1. RGB color space:             
The most popular color space in image processing.The RGB color space is an example of an additive color space: the more of each color is added, the brighter the pixel becomes and the closer it comes to white               



```python
# visualizing Red, green, blue color channel seperately
import cv2
import numpy
import matplotlib.pyplot as plt

# reading the image
bgr_image = cv2.imread('./data/rgb.jpg')  # opencv always represents image in BGR instead of RGB

# convert to RGB color image
rgb_image= cv2.cvtColor(bgr_image, cv2.COLOR_BGR2RGB)

# printing the shape of the image
print("RGB space image which has shape of {} contains {} channels".format(bgr_image.shape, bgr_image.shape[2]))

# splitting each channel to visualize each color space
red = rgb_image[:,:,0]  # blue channel
green =rgb_image[:,:,1]  # green channel
blue = rgb_image[:,:,2]  # red channel 


# plotting each color channel seperately
fig = plt.figure(figsize = (12,12))

ax1 = fig.add_subplot(1,4,1)
ax1.set_title("Original Image")
ax1.axis("off")
ax1.imshow(rgb_image)


ax2 = fig.add_subplot(1,4,2)
ax2.set_title("Red channel")
ax2.axis("off")
ax2.imshow(red)

ax3 = fig.add_subplot(1,4,3)
ax3.set_title("Green channel")
ax3.axis('off')
ax3.imshow(green)

ax4 = fig.add_subplot(1,4,4)
ax4.set_title("Blue channel")
ax4.axis('off')
ax4.imshow(blue)

plt.show()
```

#### 2. HSV color space:                        
It is mostly used for color segmentation purpose.   

Hue: Which “pure” color we are examining. For example, all shadows and tones of the color “red” will have the same Hue.         

Saturation: How “white” the color is. A fully saturated color would be “pure,” as in “pure red.” And a color with zero saturation would be pure white.             

Value: The Value allows us to control the lightness of our color. A Value of zero would indicate pure black, whereas increasing the value would produce lighter colors.


```python
# conversion of HSV color space
import cv2
import numpy as np
import matplotlib.pyplot as plt

# reading the image
bgr_image = cv2.imread('./data/hari_1.png')

# convert to RGB
rgb_image = cv2.cvtColor(bgr_image, cv2.COLOR_BGR2RGB)

# converting to HSV
hsv_image = cv2.cvtColor(bgr_image, cv2.COLOR_BGR2HSV)

# printing the shape of the image
print("HSV space image which has shape of {} contains {} channels".format(bgr_image.shape, bgr_image.shape[2]))

# splitting each channel to visualize each color space
Hue = hsv_image[:,:,0]  # Hue channel
Saturation =hsv_image[:,:,1]  # Saturation channel
Value = hsv_image[:,:,2]  # Value channel 


# plotting each color channel seperately
fig = plt.figure(figsize = (12,12))

ax1 = fig.add_subplot(1,4,1)
ax1.set_title("Original Image")
ax1.axis("off")
ax1.imshow(rgb_image)

ax2 = fig.add_subplot(1,4,2)
ax2.set_title("Hue")
ax2.axis("off")
ax2.imshow(Hue)


ax3 = fig.add_subplot(1,4,3)
ax3.set_title("Saturation")
ax3.axis("off")
ax3.imshow(Saturation)

ax4 = fig.add_subplot(1,4,4)
ax4.set_title("Value")
ax4.axis('off')
ax4.imshow(Value)

plt.show()
```

 the value controls the actual lightness of our color, while both Hue and Saturation define the actual color and shade.

#### 3. LAB color space:              
The Euclidean distance between two arbitrary colors in the L*a*b* color space has actual perceptual meaning.due to the distance between colors having an actual perceptual meaning, allowing us to overcome various lighting condition problems.          

L-channel: The “lightness” of the pixel. This value goes up and down the vertical axis, white to black, with neutral grays at the center of the axis.             

a-channel: Originates from the center of the L-channel and defines pure green on one end of the spectrum and pure red on the other.                 

b-channel: Also originates from the center of the L-channel, but is perpendicular to the a-channel. The b-channel defines pure blue at one of the spectrum and pure yellow at the other.

The a* and b* then determine the shade and color of the pixel.


```python
# conversion of LAB color space
import cv2
import numpy as np
import matplotlib.pyplot as plt

# reading the image
bgr_image = cv2.imread('./data/hari_1.png')

# convert to RGB
rgb_image = cv2.cvtColor(bgr_image, cv2.COLOR_BGR2RGB)

# converting to HSV
lab_image = cv2.cvtColor(bgr_image, cv2.COLOR_BGR2LAB)

# printing the shape of the image
print("LAB space image which has shape of {} contains {} channels".format(bgr_image.shape, bgr_image.shape[2]))

# splitting each channel to visualize each color space
Lightness = lab_image[:,:,0]  # lightness channel (0 to 255)
a = lab_image[:,:,1]  # a channel
b = lab_image[:,:,2]  # b channel 


# plotting each color channel seperately
fig = plt.figure(figsize = (12,12))

ax1 = fig.add_subplot(1,4,1)
ax1.set_title("Original Image")
ax1.axis("off")
ax1.imshow(rgb_image)

ax2 = fig.add_subplot(1,4,2)
ax2.set_title("Lightness")
ax2.axis("off")
ax2.imshow(Lightness)


ax3 = fig.add_subplot(1,4,3)
ax3.set_title("a channel")
ax3.axis("off")
ax3.imshow(a)

ax4 = fig.add_subplot(1,4,4)
ax4.set_title("b channel")
ax4.axis('off')
ax4.imshow(b)

plt.show()
```
