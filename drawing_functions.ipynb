## Drawing functions:              
The notebook provides a guide for all the drawing functions present in OpenCV. The user could draw shapes like line,circle,rectangle,Polygons, triangles and complex shapes are drawn by making seperate function.There are some additional functions available in opencv which is very useful for many applications. These works with respect to the image coordinates.

```python
# import library
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# creating an empty RGB image using numpy with shape of 500*500
null_img = np.zeros(shape= (500,500,3), dtype= np.uint8)

# plotting the empty image
plt.imshow(null_img)
plt.show()
```

#### 1. ArrowLine, Line, Circle, Rectangle, Ellipse:
ArrowLine - Draws an arrow segment pointing from the first point to the second one.                
Line - Draws a line pointing from the first point to the second one.                     
Circle - Draws a circle pointing from center axis with respect of radius                        
Rectangle - Draws a rectangle outline or a filled rectangle whose two opposite corners are pt1 and pt2.

```python
# importing library
import cv2
import numpy as np
import matplotlib.pyplot as plt


# empty image
img_drawn= np.zeros(shape= (500,500,3), dtype= np.uint8)

# image arrow drawn
img_drawn= cv2.arrowedLine(img= img_drawn, pt1= (100,100), pt2= (300,30), color= (255,255,255),
                           thickness= 2, shift= 0, tipLength= 0.2) #white


# Image Line drawn
img_drawn = cv2.line(img= img_drawn, pt1= (100,150), pt2= (300,75), color= (0,0,255),
                     thickness= 2, lineType= cv2.LINE_AA, shift= 0) # blue

# Image Circle drawn
img_drawn = cv2.circle(img= img_drawn, center= (200,89), radius= 21, color= (0,255,255),
                       thickness= -1, lineType= cv2.LINE_AA, shift= 0) # blue+green= cyan

# Image Rectangle drawn
img_drawn = cv2.rectangle(img= img_drawn, pt1= (130,100), pt2= (150,30), color= (255,255,0), thickness= -1) #red+green = Yellow

# Image Ellipse drawn
img_drawn = cv2.ellipse(img= img_drawn, center= (250,250), axes= (150,50), angle= 30, startAngle= 0, endAngle= 360, 
                        color= (255,0,0), thickness= 5)

# plotting the image
plt.figure(figsize= (8,8))
plt.title("Arrowed Line,Line,Circle,Rectangle,Ellipse")
plt.imshow(img_drawn)
plt.show()
```

```python
# slicing / zooming the image to visualize 
plt.imshow(img_drawn[9:160, 50:320, :])
```

#### 2. ClipLine:
calculates a part of the line segment that is entirely within the specified rectangle. It returns false if the line segment is completely outside the rectangle. Otherwise, it returns true .

```python
# importing library
import cv2
import numpy as np
import matplotlib.pyplot as plt

# empty image
out_img = np.zeros(shape= (500,500,3), dtype= np.uint8)

# drawing both rectangle and line
out_img= cv2.rectangle(img= out_img, pt1= (100,100), pt2= (300,300), color= (0,255,0), thickness = 3) # green
out_img= cv2.line(img= out_img, pt1= (200,0), pt2= (200,500), color= (255,0,255), thickness= 3) # blue+red= magnetta

# Clipline imgRect : pt1,pt2 of rectangle, pt1 : pt1 of line, pt2 : pt2 of line

ret, p1,p2= cv2.clipLine(imgRect= (100,100,300,300), pt1= (200,0), pt2= (200,300))

print(ret)
if ret:
    cv2.line(out_img, p1, p2, color= (255,255,0), thickness= 3)

plt.imshow(out_img)
plt.show()
```

#### 3. Polylines:
Draws one or more polygonal curves.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt


# empty iimage
polygon_img= np.zeros(shape= (500,500,3), dtype= np.uint8)

pts1= np.array([[25, 70], [25, 145],
                [75, 190], [150, 190],
                [200, 145], [200, 70], 
                [150, 25], [75, 25]],
               np.int32)
pts2 = np.array([[125, 170], [125, 260], 
                [210, 300], [300, 260], 
                [300, 170], [210, 120]],
               np.int32)
# polygon
polygon_img= cv2.polylines(img= polygon_img, pts= [pts1,pts2], 
                           isClosed= True, color= (255,0,0), thickness= 6) #red

# plotting the image
plt.title("Polygon shape : Used for shape detection/Segmentation")
plt.imshow(polygon_img)
plt.show()
```

#### 4. fillPoly:
fills an area bounded by several polygonal contours. The function can fill complex areas, for example, areas with holes, contours with self-intersections (some of their parts), and so forth

```python
import cv2 
import numpy as np
import matplotlib.pyplot as plt

# empy image
polygon_img= np.zeros(shape= (500,500,3), dtype= np.uint8)



pts = np.array([[125, 170], [125, 260], 
                [210, 300], [300, 260], 
                [300, 170], [210, 120]],
               np.int32)

#fill poly
polygon_img = cv2.fillPoly(img= polygon_img, pts= [pts], color= (0,255,255)) #cyan

#plt
plt.title("FillPoly")
plt.imshow(polygon_img)
plt.show()
```

#### 5. fillConvexPoly:
draws a filled convex polygon. This function is much faster than the function fillPoly . It can fill not only convex polygons but any monotonic polygon without self-intersections, that is, a polygon whose contour intersects every horizontal line (scan line) twice at the most (though, its top-most and/or the bottom edge could be horizontal).

```python

import cv2
import numpy as np
import matplotlib.pyplot as plt


# empty iimage
polygon_img= np.zeros(shape= (500,500,3), dtype= np.uint8)

# triangle coordinates
pts= np.array([[100,300],[200,100],[150,350]])

# polygon
polygon_img= cv2.fillConvexPoly(img= polygon_img, points= pts, color= (0,255,0)) #Yellow

# plotting the image
plt.title("Polygon shape : Efficient than fillPoly")
plt.imshow(polygon_img)
plt.show()
```

#### 6. ellipse2Poly:
The function ellipse2Poly computes the vertices of a polyline that approximates the specified elliptic arc

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

# empty image
out_img= np.zeros(shape= (500,500,3), dtype= np.uint8)

# points for drawing the polygon
pts = cv2.ellipse2Poly(center= (250,250), axes= (100,50), angle= 20, arcStart= 0, arcEnd= 180, delta= 1) 

# drawing the convexpoly
out_image= cv2.fillConvexPoly(img= out_img, points= pts, color= (255,255,255)) #white like moon

# plot
plt.title("points from ellipse2Poly")
plt.imshow(out_image)
plt.show()
```

#### 7. drawMarker:

Draws a marker on a predefined position in an image.There are several marker types available predetermined in opencv

```python
import cv2
import numpy as np
import  matplotlib.pyplot as plt

# empt image
markers_img= np.zeros(shape= (500,500,3), dtype= np.uint8)

# markwrs
markers_img= cv2.drawMarker(img= markers_img, position= (250,250), color= (255,255,255), markerType= cv2.MARKER_CROSS, markerSize= 30, thickness= 5)
markers_img= cv2.drawMarker(img= markers_img, position= (100,100), color= (255,0,255), markerType= cv2.MARKER_TILTED_CROSS, markerSize= 30, thickness= 5)
markers_img= cv2.drawMarker(img= markers_img, position= (400,400), color= (0,255,255), markerType= cv2.MARKER_STAR, markerSize= 30, thickness= 5)
markers_img= cv2.drawMarker(img= markers_img, position= (150,150), color= (255,255,0), markerType= cv2.MARKER_DIAMOND, markerSize= 30, thickness= 5)
markers_img= cv2.drawMarker(img= markers_img, position= (50,50), color= (255,0,0), markerType= cv2.MARKER_SQUARE, markerSize= 30, thickness= 5)
markers_img= cv2.drawMarker(img= markers_img, position= (350,350), color= (0,0,255), markerType= cv2.MARKER_TRIANGLE_UP, markerSize= 30, thickness= 5)
markers_img= cv2.drawMarker(img= markers_img, position= (450,450), color= (0,255,0), markerType= cv2.MARKER_TRIANGLE_DOWN, markerSize= 30, thickness= 5)

plt.imshow(markers_img)
plt.show()
```
